```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder


columns_to_drop = ['ID', 'DEVICETYPE', 'PLATFORM_ID', 'CHANNELTYPE', 'USERAGENT',
                   'BIDREQUESTIP', 'USERPLATFORMUID', 'URL']


test_df.drop(columns=columns_to_drop, inplace=True)
```

```python
test_df['USERCITY'] = test_df['USERCITY'].fillna(test_df.groupby('USERZIPCODE')['USERCITY'].transform('first'))


all_keywords = test_df['KEYWORDS'].str.cat(sep='|')
word_list = list(set(all_keywords.split('|')))


word_list = list(filter(None, word_list))


for word in word_list:
    test_df[word] = test_df['KEYWORDS'].apply(lambda x: 1 if word in x else 0)


columns_to_drop = ['KEYWORDS']
test_df.drop(columns_to_drop, axis=1, inplace=True)


encoder = OneHotEncoder(sparse=False)
platform_type = test_df['PLATFORMTYPE'].values.reshape(-1, 1)
encoded_features = encoder.fit_transform(platform_type)
encoded_df = pd.DataFrame(encoded_features, columns=encoder.categories_[0])


test_df = pd.concat([test_df, encoded_df], axis=1)


test_df.drop('PLATFORMTYPE', axis=1, inplace=True)


test_df.dropna(subset=['USERCITY', 'USERZIPCODE'], inplace=True)


city_encoder = LabelEncoder()
test_df['USERCITY'] = city_encoder.fit_transform(test_df['USERCITY'])

zipcode_encoder = LabelEncoder()
test_df['USERZIPCODE'] = zipcode_encoder.fit_transform(test_df['USERZIPCODE'])

# Print the updated DataFrame
print(test_df.head(5))
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from math import sqrt


X_train = train_df.drop('IS_HCP', axis=1)
y_train = train_df['IS_HCP']


X_test = test_df.drop('IS_HCP', axis=1)

y_test = test_df['IS_HCP']


model = xgb.XGBRegressor()


model.fit(X_train, y_train)


y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE): {:.2f}".format(mse))


rmse = sqrt(mse)
print("Root Mean Squared Error (RMSE): {:.2f}".format(rmse))
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score



X = train_df.drop('IS_HCP', axis=1)
y = train_df['IS_HCP']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)


model = xgb.XGBClassifier()


model.fit(X_train, y_train)


y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
